```python
#import libraries and read files that contain the reward per episode data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data from CSV files
ppo_data = pd.read_csv('PPO.csv')
td3_data = pd.read_csv('TD3.csv')
a2c_data = pd.read_csv('A2C.csv')
ddpg_data = pd.read_csv('DDPG.csv')


# Add a column to identify the method
ppo_data['method'] = 'PPO'
td3_data['method'] = 'TD3'
a2c_data['method'] = 'A2C'
ddpg_data['method'] = 'DDPG'

# Filter the data to include only the first 350 episodes
ppo_data = ppo_data[ppo_data['Episode'] <= 350]
td3_data = td3_data[td3_data['Episode'] <= 350]
a2c_data = a2c_data[a2c_data['Episode'] <= 350]
ddpg_data = ddpg_data[ddpg_data['Episode'] <= 350]

# Concatenate all data into a single DataFrame
all_data = pd.concat([ppo_data, a2c_data, ddpg_data, td3_data])

# Plotting
plt.figure(figsize=(12, 8))
sns.lineplot(data=all_data, x='Episode', y='Rewards', hue='method', palette='tab10')

plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Reward per Episode for Different Methods (First 350 Episodes)')
plt.legend(title='Method')
plt.grid(True)

plt.show()
```
